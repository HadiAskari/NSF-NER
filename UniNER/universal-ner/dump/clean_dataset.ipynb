```python
import pandas as pd
from ast import literal_eval
```

```python
df=pd.read_csv('test_dataset.csv')
```

```python
df.head()
```

```python
wanted=['Application','Device','Error_Name','Library','Operating_System']
```

```python
text_list=df['Text'].to_list()
entity_list=df['entities'].to_list()
entity_types_list=df['entity_types'].to_list()
```

```python
for k,v in enumerate(entity_list):
    entity_list[k]=literal_eval(v)
```

```python
for k,v in enumerate(entity_types_list):
    entity_types_list[k]=literal_eval(v)
```

```python
new_entities=[]
new_types=[]
for entities, types in zip(entity_list,entity_types_list):
    for entity, type in zip(entities,types):
        if type in wanted:
            continue
        else:
            entities.remove(entity)
            types.remove(type)
    new_entities.append(entities)
    new_types.append(types)
```

```python
entity_types_list
```

```python
new_types
```

```python
dic={'Text':text_list,'entities':new_entities, 'entity_types':new_types}
```

```python
df_new=pd.DataFrame.from_dict(dic)
```

```python
df_new.to_csv('selected_test_dataset.csv',index=False)
```
