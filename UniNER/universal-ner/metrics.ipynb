```python
import pandas as pd 
from ast import literal_eval
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from collections import defaultdict
import re
from sklearn.preprocessing import MultiLabelBinarizer
import sklearn
```

```python
df=pd.read_csv('SecData_Vulnerability_combined.csv')
```

```python
df.head
```

```python
df.head()
```

```python
truth=[]
predictions=[]

for i in df.itertuples():
    truth.append(literal_eval(i[2]))
    # entity_types.append(literal_eval(i[4]))
    
    p = re.compile(r"^defaultdict\(<class '(\w+)'>")
    c = p.findall(i[4])[0]
    # print(c)

    new_s = i[4].replace("<class '%s'>"% c, c)
    # print(new_s)
    new_s=eval(new_s)
    print(new_s)
    count=0
    temp=[]
    for k,v in new_s.items():
        try:
            temp.extend(literal_eval(v[0]))
            count+=1
        except:
            temp.extend([])
    predictions.append(temp)
    
    



# literal_eval(new_s)
```

```python
truth= [list(set(x)) for x in truth]
```

```python
len(truth)
```

```python
len(predictions)
```

```python
truth
```

```python
predictions
```

```python
total_classes=[]
positive_classes=[]
for item in truth:
    if not item:
        continue
    else:
        for vals in item:
            total_classes.append(vals)
            positive_classes.append(vals)
for item in predictions:
    if not item:
        continue
    else:
        for vals in item:
            total_classes.append(vals)
```

```python
total_classes=list(set(total_classes))
```

```python
positive_classes=list(set(positive_classes))
```

```python
len(total_classes)
```

```python
len(positive_classes)
```

```python
mlb = MultiLabelBinarizer()
```

```python
mlb.fit([positive_classes])
```

```python
truth_binarized= mlb.transform(truth)
```

```python
truth_binarized.shape
```

```python
len(mlb.classes_)
```

```python
predictions_binarized=mlb.transform(predictions)
```

```python
len(mlb.classes_)
```

```python
predictions_binarized.shape
```

```python
f1 = f1_score(truth_binarized, predictions_binarized, average='micro')
f1
```

```python
recall = recall_score(truth_binarized, predictions_binarized, average='micro')
recall
```

```python
precision = precision_score(truth_binarized, predictions_binarized, average='micro')
precision
```

```python
total=0
acc=0
for tru, pred in zip(truth, predictions):
    total+=len(tru)
    for items in tru:
        if items in pred:
            acc+=1
    
```
